```python
import requests
import json
```

# Ontology Index

```python
CELLXGENE_FILTERS_URL = "https://api.cellxgene.cziscience.com/wmg/v2/filters"
CELLXGENE_PRIMARY_FILTER_URL = "https://api.cellxgene.cziscience.com/wmg/v2/primary_filter_dimensions"
OUTPUT_FILENAME = "ontology_index.json"
payload_filter = {
        "filter": {
            "organism_ontology_term_id": "NCBITaxon:9606"
        }
    }
```

```python
response = requests.post(CELLXGENE_FILTERS_URL, json=payload_filter)
```

```python
api_data = response.json()
```

```python
tissues = [
            {"id": key, "name": value} 
            for ont_dict in api_data.get("filter_dims", {}).get("tissue_terms", [])
            for key, value in ont_dict.items()
        ]
```

```python
cells = [
            {"id": key, "name": value} 
            for ont_dict in api_data.get("filter_dims", {}).get("cell_type_terms", [])
            for key, value in ont_dict.items()
        ]
```

```python
resp = requests.get(CELLXGENE_PRIMARY_FILTER_URL)
primary_data = resp.json()
```

```python
primary_data['gene_terms']['NCBITaxon:9606']
```

```python
genes = [
            {"id": key, "name": value} 
            for ont_dict in primary_data.get('gene_terms', {}).get('NCBITaxon:9606', [])
            for key, value in ont_dict.items()
        ]
```

```python
ontology_index = {
            "tissues": tissues,
            "cell_types": cells,
            "genes": genes
        }
```

```python
with open(OUTPUT_FILENAME, 'w') as f:
            json.dump(ontology_index, f, indent=2)
```
